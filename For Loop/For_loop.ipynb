# enumerate function 

```python
# enumerate function --> print indexs and output object
```

```python
# exapmle 
word='abcde'
for item in enumerate(word):
        print(item)
```

```python
word='abcde'
for item,letter in enumerate(word):
        print(item)
        print(letter)
        print('\n')
```

# Zip Function

```python
# zip function --> mixed two list 
```

```python
# example
mylist1=[1,2,3,4,4]
mylist2=['a','b','c','d','e']
```

```python
zip(mylist1,mylist2)
```
