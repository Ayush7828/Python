```python
# create vehicle class with instance variable
```

```python
class Vehicle:
    def __init__(self,name,speed,average):
        self.name=name
        self.speed=speed
        self.average=average
        
    def sitting_capacity(self,capacity):
        print("Car name is {} {}".format(self,name,capacity))
```

```python
tata=Vehicle('Nano',120,20.0)
```

```python

tata.average
```
