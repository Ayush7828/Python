```python
my_dog=Dog('Huskie')
```

```python
print(my_dog.breed)

my_dog.bark()
```

```python
class Circle:
    # class circle attribute
    pie=3.14
    
    def __init__(self,radius,area):
        # instance variable
        self.radius=radius
        self.area=radius*radius*self.pie
        
    def calculate(self):
        return self.radius*self.pie*2
    
```

```python
value=Circle(20,40)
```

```python

value.calculate()
```

```python
value.area
```
