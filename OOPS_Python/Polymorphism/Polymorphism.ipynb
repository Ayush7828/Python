```python
class Dog:
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        return self.name  +" says   Wooof!"
```

```python
class Cat:
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        return self.name+ " says  Meow!"
```

```python
niko=Dog('niko')
felix=Cat('felix')
```

```python
niko.speak()
```

```python
felix.speak()
```
