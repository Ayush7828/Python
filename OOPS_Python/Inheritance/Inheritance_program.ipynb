```python
# single level Inheritance
```

```python
class Animal:
    def speaking(self):
        print("Animal Speaking")
    
class Dog(Animal):
    def bark(self):
        print("Dog Barking Woof")
        
```

```python
my_dog=Dog()
```

```python
my_dog.bark()
my_dog.speaking()
```

# Multilevel Inheritance

```python
# class class1:  
#     <class-suite>   
# class class2(class1):  
#     <class suite>  
# class class3(class2):  
#     <class suite>
```

```python
class Animal:
    def speak(self):
        print("Animal Speaking")
        
        
class Dog(Animal):
    def barking(self):
        print("Dog Barking woof")

class DogChild(Dog):
    def eat(self):
        print("Dog Child Eating")
        
```

```python
my_dog=DogChild()
```

```python
my_dog.eat()
```

```python
my_dog.eat()
my_dog.speak()
my_dog.barking()
```

# Multiple Inheritance

```python
# class Base1:  
#     <class-suite>  
  
# class Base2:  
#     <class-suite>  
# .  
# .  
# .  
# class BaseN:  
#     <class-suite>  
  
# class Derived(Base1, Base2, ...... BaseN):  
# #     <class-suite>  
```

```python
class Calculation1:
    def addition(self,a,b):
        return a+b
    
    
    
    
class Calculation2:
    def multiply(self,a,b):
        return a*b
    

class Derived(Calculation1,Calculation2):
    def thanku(self):
        print("Thankuu ")
    
```

```python
number=Derived()
```

```python
number.addition(25,50.4)
```

```python
number.multiply(21,86.1)
```

```python
number.thanku()
```
