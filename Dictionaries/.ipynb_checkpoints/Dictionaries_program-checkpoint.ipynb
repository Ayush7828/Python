```python
# access item based on the key
```

```python
my_disc={"car":'Audi', 'Car1':'BMW','Car2':'Tata'}
```

```python
my_disc['car']
```

```python
# use loop to print
# output show keys
for i in my_disc:
    print(i)
```

```python
for i in my_disc.values():
    print(i)
```

```python
# both are prints like key and value 
for i in my_disc.items():
    print(i)
```

```python
# adding item in dicitinoraies 
my_disc['car4']='Honda City'
```

```python
print(my_disc)
```

# Nested Dictionaries

```python
car1_model={'Mercedes':1960}
car2_model={'Audi':1970}
car3_model={'Tata-Motors':2000}


car_models={'car1':car1_model,'car2':car2_model,'car3':car3_model}
```

```python
print(car_models)
```

```python
# accessing the item in nested dictionaries
print(car_models['car1'])
print(car_models['car3'])
```

```python
# only value print 
# like this 
print(car_models['car1']['Mercedes'])
print(car_models['car2']['Audi'])
print(car_models['car3']['Tata-Motors'])
```
