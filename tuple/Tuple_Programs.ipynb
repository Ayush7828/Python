```python
# tuple
# Python Tuple is used to store the sequence of immutable Python objects. 
# The tuple is similar to lists since the value of the items stored in the 
# list can be changed, whereas the tuple is immutable, and the value of the items stored in the tuple cannot be changed.
```

```python
# create empty typle
my_tuple=()
```

```python
print(my_tuple)
type(my_tuple)
```

```python
my_tuple1=(10,20,30,40,40,50)
```

```python
print(my_tuple1)
```

```python
# use index print typle
```

```python
print(my_tuple1[0])
print(my_tuple1[1])
print(my_tuple1[2])
print(my_tuple1[3])
```

```python
#slicing tuple
```

```python
print(my_tuple1[0:])
print(my_tuple1[1:])
print(my_tuple1[2:])
```

```python
print(my_tuple1[:-2])
print(my_tuple1[1:3])
print(my_tuple1[2:5])
```
