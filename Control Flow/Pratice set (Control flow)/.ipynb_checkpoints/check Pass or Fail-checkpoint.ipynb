```python
print("Enter 5 subject marks ")
english=int(input("English"))
maths=int(input("maths "))
hindi=int(input("Hindi "))
physics=int(input("Physics "))
chemistry=int(input("Chemistry "))
add=english+maths+hindi+physics+chemistry
print(add)
print(add/5.0)
if add>40:
    print("Pass ",add)
else:
    print("Fail ",add)
```
