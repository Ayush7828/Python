```python
# check area and perimeter which one is greater
```

```python
length=int(input("Enter a length "))
breadth=int(input("Enter a breadth "))
area=length*breadth
perimeter=2*(length+breadth)
print("Area = ",area)
print("Perimeter = ",perimeter)

if area>perimeter:
        print("area is greater ",area," ",perimeter)
else: 
        print("Perimeter is greater ",perimeter)
```
