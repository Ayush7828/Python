```python
# if , else statement 
# syntax -->  if some_condition:
#           execute code
#             else:
#             execute code
```

```python
age=int(input("Enter a age "))
if age>18:
    print("You are eligible for give the vote")
else:
    print("You a kid")
```

```python
#example 
char =str(input("Enter a character "))
if char>'a'and char<'z':
    print("it is a Lowercase Character ",char)
elif char>'A'and char<'Z':
    print("It is Uppercase Character ",char)
else:
    print("Invaild ",char)
```
