# String Formating And For Printing

```python
#Two Methods to format String
# 1) .format()
# 2) f-String(Format String Literal)
```

# 1) Formating String    .format()

```python
# A good way to format Object into your String for print Statement is with the String .format() 
# method.
# Syntax ->  'String here{} then also' .format('something1 ','something 2')
```

```python
print('Hello {}'.format('Ayush'))
```

```python
print("The {} {} {} {}".format('fox','Snake','Tiger','.'))
```

```python
#insert in index form
print("The {0} {0} {0} {3}".format('fox','Snake','Tiger','.'))
```

```python
#use variable method to format String
print("The {s} {T} {f}".format(f='fox',s= 'Snake',T='Tiger'))
```

# Float Formating follows "{value:Width.percision f}" 

```python
result=100/777
```

```python
result
```

```python
print("The Result is {r:1.3f}".format(r=result))
```

```python
# print white space because calculate the white spaces 
print("The Result is {r:10.3f}".format(r=result))
```
