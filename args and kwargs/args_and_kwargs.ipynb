```python
def my_fun1(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit choice is {}'.format(kwargs['fruit']))
    else:
        print("I did not find any Fruit here")
```

```python
my_fun1(fruit='apple')
```
